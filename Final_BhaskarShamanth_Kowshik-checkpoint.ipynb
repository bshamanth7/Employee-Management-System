```python
#define your own Full Time Employee class here
#Full Time Employee class inherits from Employee class
#the order of the four required positional arguments should be: name,address,vehicle object,salary
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class FullTimeEmployee(Employee):
    def __init__(self, name, address, vehicle, salary: float):
        Employee.__init__(self, name, address, vehicle)
        self.__salary = salary
    def get_salary(self):
        return self.__salary
    def set_salary(self, salary):
        self.__salary = salary
    def compute_compensation(self):
        return (self.__salary - get_FTE_Taxes(self.__salary))/52
    def compute_reimbursement(self, expense):
        return (get_FTE_reimbursement(expense))/52
    def __str__(self):
        emp = '\nDetails of this Full Time Employee are: \n'
        emp+= 'Employee Name: ' + self.get_emp_name() + '; Employee Address: ' + self.get_emp_addr()+'\n'
        emp+=  self.get_vehicle().__str__()+'\n'
        emp+= 'Salary: {0:0.2f}'.format(self.__salary)
        return(emp)

    
def get_FTE_Taxes(salary):
    tax_dict = {'s1': 0.18, 's2': 0.28, 's3': 0.33}
    if salary < 25000:
        tax = salary*tax_dict['s1']
    elif 25000< salary <= 55000:
        tax = (25000*tax_dict['s1']) + ((salary-25000)*tax_dict['s2'])
    elif salary > 55000:
        tax = (25000*tax_dict['s1']) + (30000*tax_dict['s2']) + ((salary-55000)*tax_dict['s3'])
    else:
        tax = 0
    return(tax)

def get_FTE_reimbursement(expense):
    if expense <= 10000:
        reim = expense
    else:
        reim = 10000 + (0.5* (expense-10000))
    return(reim)    
```

```python
#define your own Hourly Employee class here
#Hourly Employee class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,hourly rate
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class HourlyEmployee(Employee):
    def __init__(self, name, address, vehicle, hours_worked: int, hourly_rate: float):
        Employee.__init__(self, name, address, vehicle)        
        self.__hours_wrkd = hours_worked
        self.__hourly_rate = hourly_rate

    def get_hours_worked(self):
        return self.__hours_wrkd
    def set_hours_worked(self, hours_worked):
        self.__hours_wrkd = hours_worked
    def get_hourly_rate(self):
        return self.__hourly_rate
    def set_hourly_rate(self, hourly_rate):
        self.__hourly_rate = hourly_rate

    def compute_compensation(self):
        if self.__hours_wrkd > 40:
            return (40* self.__hourly_rate) + ((self.__hours_wrkd-40) * (self.__hourly_rate*1.8))
        else:
            return self.__hours_wrkd * self.__hourly_rate
    def compute_reimbursement(self, expense):
        return get_HE_reimbursement(expense)
    
    def __str__(self):
        emp = '\nDetails of this Hourly Employee are:  \n'
        emp+= 'Employee Name: ' + self.get_emp_name() + '; Employee Address: ' + self.get_emp_addr()+'\n'
        emp+=  self.get_vehicle().__str__() +'\n'
        emp+= 'Hours Worked: ' + str(self.get_hours_worked())+ '; Hourly Rate: ' + str(self.get_hourly_rate())
        return(emp)

def get_HE_reimbursement(expense):
    if expense <= 100:
        reim = expense
    else:
        reim = 100
    return(reim)
```

```python
#define your own Consultant class here
#Consultant class inherits from Employee class
#the order of the five required positional arguments should be: name,address,vehicle object,hours worked,project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

from typing import Literal
class Consultant(Employee):
    def __init__(self, name, address, vehicle, hours_worked: int, project_type:Literal[1,2,3]):
        Employee.__init__(self, name, address, vehicle)
        self.__hours_wrkd = hours_worked
        self.__project_type = project_type
        
    def get_hours_worked(self):
        return self.__hours_wrkd
    def set_hours_worked(self, hours_worked):
        self.__hours_wrkd = hours_worked
    def get_project_type(self):
        return self.__project_type
    def set_project_type(self, project_type):
        self.__project_type = project_type 
    
    def compute_compensation(self):
        return self.__hours_wrkd * get_CE_hourly_rate(self.__project_type)
    def compute_reimbursement(self, expense):
        return(get_CE_reimbursement(expense, self.__project_type))
    
    def __str__(self):
        emp =  '\nDetails of this Consultant are: \n'
        emp+=  'Employee Name: ' + self.get_emp_name() + '; Employee Address: ' + self.get_emp_addr()+ '\n'
        emp+=  self.get_vehicle().__str__() +'\n'
        emp+=  'Hours Worked: ' + str(self.get_hours_worked())+ '; Project Type: ' + str(self.get_project_type()) 
        return(emp)

def get_CE_hourly_rate(project_type):
    hr_dict = {1: 55, 2: 70, 3: 85}
    return hr_dict[project_type]

def get_CE_reimbursement(expense, project_type):
    ce_reim_dict = {1: 1, 2: 0.9, 3: 0.8}
    reim = expense*ce_reim_dict[project_type]
    return(reim)
```

```python
#define your own Management class here
#Management class inherits from both Full Time Employee class and Consultant class
#the order of the six required positional arguments should be: name, address,vehicle object, salary, hours worked, project type
#please use the conventional way to define function names and refer to the following test to adjust function names
#see detailed requirements on compute_compensation() function and compute_reimbursement() function in the pdf file
#no need to handle exceptions here

class Management(FullTimeEmployee,Consultant):
    def __init__(self, name, address, vehicle, salary: float, hours_worked, project_type):
        Employee.__init__(self, name, address, vehicle,)
        FullTimeEmployee.__init__(self, name, address, vehicle, salary)
        Consultant.__init__(self, name, address, vehicle, hours_worked, project_type)
   
    def compute_compensation(self):
        FTE_comp = (self.get_salary() - get_FTE_Taxes(self.get_salary()))/52
        CE_comp = (self.get_hours_worked() * get_CE_hourly_rate(self.get_project_type()))
        return(FTE_comp + CE_comp)
    def compute_reimbursement(self,Annualexpense, Weeklyexpense):
        FTE_reim = (get_FTE_reimbursement(Annualexpense))/52
        CE_reim = (get_CE_reimbursement(Weeklyexpense, self.get_project_type()))
        return(FTE_reim + CE_reim)
    
    def __str__(self):
        emp =  '\nDetails of this Management are: \n'
        emp+=  'Employee Name: ' + self.get_emp_name() + '; Employee Address: ' + self.get_emp_addr()+ '\n'
        emp+=   self.get_vehicle().__str__() +'\n'
        emp+=  'Salary: {0:0.2f}'.format(self.get_salary())+ '; Hours Worked: ' + str(self.get_hours_worked())+ '; Project Type: ' + str(self.get_project_type()) 
        return(emp)
```

```python
#complete this get_vehicle_input() function to prepare for the main menu application
#this function is used to ask for vehicle information and then return them: make,model,year,mileage
#this function doesn't take arguement(s)
#no need to handle exceptions of make input and model input
#handle exceptions to make sure year is a 4-digit positive number between 1900 and 2023
#handle exceptions to make sure mileage is a positive number
def get_vehicle_input():
    veh_make = input('Enter the vehicle make: ')
    veh_model = input('Enter the vehicle model: ')
    while True:
        try:
            veh_year = int(input('Enter the year of manufacture (yyyy): '))
            if veh_year < 1900 or veh_year > 2023 or len(str(veh_year)) != 4:
                raise ValueError
            break
        except ValueError:
            print("\nPlease enter an integer value for year in the format of yyyy between 1900 and 2023.")

    while True:
        try:
            veh_mileage = int(input("Enter the mileage: "))
            if veh_mileage < 0:
                raise ValueError
            break
        except ValueError:
            print("\nPlease enter a positive number value for mileage.") 
    return (veh_make, veh_model, veh_year, veh_mileage)
```

```python
#complete this read_file_data() function to prepare for the main menu application
#this function is used to read information from an existing databse/binary file, which is empdata.dat in the final exam
#this function doesn't take argument(s)
#this function returns a list containing all employee objects stored in the database; each employ object is an element in the list; the list will also be used and updated later
#need to handle exceptions if file doesn't exist
import pickle
def read_file_data():
    try:
        with open('empdata.dat', 'rb') as file:
            employees=[]
            while(True):
                try:
                    emp_record = pickle.load(file)
                    employees.append(emp_record)
                except EOFError:
                    break
    except FileNotFoundError:
        print("Error: File not found.")
        employees = []
    return employees
```

```python
#complete option 5 related code here
#the function is used to show some basic statistics. They are: 
#(1) the number of employees stored in the database
#(2) the highest weekly compensation
#(3) the mean weekly compensation
#(4) the number of employees who have a vehicle with over 100,000 mileage
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions

def run_option5(my_lst):
    
    print("\n============================================================")        
    print("Below is the statistics of all employees in this database.")
    print("==============================================================")
    #computation
    hc = 0
    sumT = 0
    milcount = 0
    for i in range(0, len(my_lst)):
        hc = max(hc, my_lst[i].compute_compensation())
        sumT = sumT + my_lst[i].compute_compensation()
        if my_lst[i].get_vehicle().get_mileage() > 100000:
            milcount+=1
    mc = sumT/len(my_lst)
    
    print('There are ' + str(len(my_lst)) + ' emplyees stored in this employee database.')
    print('The highist weekly compensation is ${0:0.2f}.'.format(hc))
    print('The mean weekly compensation is: ${0:0.2f}.'.format(mc))
    print('The number of employees who have a vehicle with over 100,000 mileage is ' + str(milcount)+'.')
```

```python
#complete option 7 related code here
#the function is used to store changes into a new file/database (i.e. empdata_updated.dat) and exit the program
#the first step is to double check with the user if he/she does want to exit (not case sensitive)
#this function takes one argument, which is the emp_lst list (i.e. the list created previously which is used to store employee objects)
#this function doesn't return anything
#no need to handle exceptions

import sys
import pickle
def run_option7(my_lst):
    #open the file to write binary
    with open('empdata_updated.dat', 'wb') as fb: 
        #write the objects to the file.  
        for i in range(0, len(my_lst)):
            pickle.dump(my_lst[i], fb) # pass the necessary arguments to the pickle module's dump function.
    print('You chose to exit the program')
    sel = input('Are you sure (Y/N)?')
    if sel not in ['Y', 'N', 'y', 'n']:
        print("\nYour input doesn't match any valid option and has been considered as \'N\'.")
        print('Going back to the selection menu.')
    elif sel in ["N", "n"]:
        print("\n============================================================")        
        print("Going back to the selection menu.")
        print("==============================================================")
    else:
        print("\n============================================================")        
        print("Program successfully being closed")
        print("==============================================================")
        sys.exit()
```
